## Introduction
In this report, I will communicate the insights about the project where I needed to wrangle and analyze tweets archive from [WeRateDogs](https://twitter.com/dog_rates). 

WeRateDogs have sent their Twitter's archive until August 1, 2017, with over 5000+ tweets. Udacity filtered only those tweets that were rated and were made available in a CSV file. They also used a neural network to classify the breed of the dogs based on the images in the tweet and made them available in a file that I had to download programmatically. It was also requested that the count of retweets and favorites gathered from Twitter API.

Based on this data, it was requested to find and clean at least 8 quality issues and 2 tidiness issues. Also was requested to produce at least 3 insights and 1 visualization over the data.

## Analysis
In this analysis, I want to find the answer to the following questions:
1. Which is the most popular name? Proportionally, which is the name that has more likes and retweets? Is there a relationship between these two data?
2. Is there a source that gives higher ratings than others?
3. Does a higher rating tend to have more likes and retweets?

Let's take a closer look at each question.

### 1. Which is the most popular name? Proportionally, which is the name that has more likes and retweets? Is there a relationship between these two data?
Based on the data, the most popular names are "Lucy", "Charlie", "Cooper", "Oliver" and "Penny". I have found that "Zoey", "Earl", "Bo", "Sunny", "Gary", "Phil" and "Dexter" are the names who showed in both top 10 of names with more likes and more retweets.

Additionally, I find that there is a strong positive relationship between the likes and the retweets (as seen in the graph "Distribution of Favorites vs Retweets"), but there is a weak negative relationship between the most popular name and the names with more retweets and favorites (graphs "Distribution of Position vs Favorites" and "Distribution of Position vs Retweets". The name "Bo" is the most popular name with more retweets and favorites.







### 2. Is there a source that gives higher ratings than others?



The "Vine" and "TweetDeck" are the source that has a rating greater than the mean of the database. We can see that tweets sent by the "TweetDeck" are way less frequent than those sent by "Vine".

### 3. Does a higher rating tend to have more likes and retweets?





A higher rating tends to have a higher number of favorites and retweets, but it is not a strong relationship.
